```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")

from sklearn.preprocessing import MinMaxScaler, StandardScaler

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = [[828, 920, 1234567, 1020, 1111],
            [824, 910, 2345612, 1090, 1234],
            [880, 900, 3456123, 1010, 1000],
            [870, 990, 2312123, 1001, 1122],
            [860, 980, 3223123, 1008, 1133],
            [850, 970, 2432123, 1100, 1221]]
data = np.float32(data)
```

```python
df = pd.DataFrame(data)
df
```

```python
x_data = data[:, :-1]
y_data = data[:, [-1]]
```

```python
x_data
```

```python
scaleF = MinMaxScaler()
x_dataN = scaleF.fit_transform(x_data) #대상 데이터가 matrix여야
```

```python
x_dataN
```

```python
scaleL = MinMaxScaler()
y_dataN = scaleL.fit_transform(y_data)
```

```python
y_dataN
```

```python
#케라스로 확인
dense = Dense(units = 1, input_dim = 4)
model = Sequential( [dense] )
model.compile(loss = 'mse', optimizer = Adam(0.1) )
h = model.fit(x_dataN, y_dataN, epochs = 1000)
```

```python
h.history.keys()
```

```python
plt.plot(h.history['loss'] )
plt.show()
```

```python
w, b = model.get_weights()
w
```

```python
#첫번째 행 값으로 예측 [828, 920, 1234567, 1020]
#model.predict( [[828, 920, 1234567, 1020]] ) 가 아닌 normalize한 값으로 줘야
xn = scaleF.transform( [[828, 920, 1234567, 1020]] )
xn
result = model.predict(xn) #y라벨도 스케일링한 값이기 때문에 확인이 힘듦
```

```python
scaleL.inverse_transform(result) #역정규화
```

#### Boston 데이터셋을 정규화한 후 학습하여 첫번째 행값의 medv를 예측하시오

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

    CRIM: 범죄율 / INDUS: 비소매상업지역 면적 비율 / NOX: 일산화질소 농도 
    RM: 주택당 방 수 / LSTAT: 인구 중 하위 계층 비율 / B: 인구 중 흑인 비율 
    PTRATIO: 학생/교사 비율 / ZN: 25,000 평방피트를 초과 거주지역 비율 
    CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0 
    AGE: 1940년 이전에 건축된 주택의 비율 
    RAD: 방사형 고속도로까지의 거리 / DIS: 직업센터의 거리 / TAX: 재산세율

```python
dfBoston= pd.DataFrame(boston['data'])
dfBoston.columns = boston['feature_names']
dfBoston['medv'] = boston['target']
dfBoston
```

```python
x_data = dfBoston.iloc[:, :-1] #데이터프레임이기에 iloc 사용
y_data = dfBoston.iloc[:, [-1]]
```

```python
x_data
```

```python
scaleF = MinMaxScaler()
x_dataN = scaleF.fit_transform(x_data)
```

```python
x_dataN
```

```python
scaleL = MinMaxScaler()
y_dataN = scaleL.fit_transform(y_data)
```

```python
y_dataN
```

```python
#케라스로 확인
dense = Dense(units = 1, input_dim = 13)
model = Sequential( [dense] )
model.compile(loss = 'mse', optimizer = Adam(0.1) )
h = model.fit(x_dataN, y_dataN, epochs = 1000)
```

```python
plt.plot(h.history['loss'])
plt.show()
```

```python
w, b = model.get_weights()
w
```
