```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
df
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
#units=1 : 라벨 갯수, input_dim=1 : 특성데이터 갯수 (다중분석시 1개 이상도 가능)
dense = Dense(units=1, input_dim=1) #w, b 초기값
model = Sequential([dense])
model.compile(loss='mse', optimizer = Adam(learning_rate=0.01)) # cost function인 mse를 구현하지 않아도 기본적인 틀이 완성되어 있음.
h = model.fit(x_data, y_data, epochs = 1000)
```
