-데이터프레임의 복사

깊은 복사 vs 얕은 복사
얕은 복사란?
:복사본이나 원본이 같은 메모리번지 사용
:서로 데이터가 연결
깊은 복사란?
:복사본이나 원본이 서로 다른 메모리번지 사용
:copy() 메서드 사용

````python
data = { "name":["Haidi", "Haidi", "Haidi", "Charles", "Charles"],
        "year":[2014, 2015, 2016, 2015, 2017],
        "points":[1.5, 1.7, 3.6, 2.5, 2.9], 
        "penalty":[1, 2, 3, 4, 5] }
df5 = pd.DataFrame( data, 
                    index=["one", "two", "three", "four", "five"])
df5
````

````python
#얕은 복사 방식으로 복제본 생성
df6 = df5
df6
````

````python
#복제본에서 값 수정
df6.loc['one', 'name'] = '홍길동'
df6
````

````python
#원본도 값이 같이 수정됨
df5
````

````python
#깊은 복사 방식으로 복제본 생성
df7 = df5.copy()
df7
````

````python
df7.iloc[-1, -1] = 500
df7
````

````python
#원본은 변경 없음
df5
````
