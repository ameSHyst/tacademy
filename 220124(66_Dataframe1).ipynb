```python
from pyspark.sql.functions import when, udf, col, regexp_replace
from pyspark.sql.types import DoubleType,IntegerType, StringType
```

```python
# def fn(v):
#     if v>=10:
#         return '많음'
#     elif v>=5:
#         return '보통'
#     else:
#         return '적음'
# return fn
```

#### pandas
    df['살인 정도'] = df['살인 발생'].apply(fn)

```python
df.withColumn('살인 정도', when(df['살인 발생']>=10, 
                            '많음').when(df['살인 발생']>=5, '보통').otherwise('적음') ).toPandas()
```

```python
# ff = udf(fn) / 함수 적용 / 조건 3개 이상이면 함수 따로 만들어야
ff = udf(lambda v:'많음' if v>=10 else '적음')
df.withColumn('살인 정도1', ff(df['살인 발생'])).toPandas()
```

```python
df.withColumn('절도 발생1', regexp_replace('절도 발생', ',', '') ).toPandas()
```

```python
df = df.withColumn('절도 발생1', regexp_replace('절도 발생', ',', '') )
```

```python
df = df.withColumn('절도 발생1', df['절도 발생1'].cast(IntegerType() ) ) #string -> integer
```

```python
df = df.withColumn('절도 발생', regexp_replace('절도 발생', ',', '') ).withColumn(
'절도 검거', regexp_replace('절도 검거', ',', '') ).withColumn(
'폭력 발생', regexp_replace('폭력 발생', ',', '') ).withColumn(
'폭력 검거', regexp_replace('폭력 검거', ',', '') )
```
