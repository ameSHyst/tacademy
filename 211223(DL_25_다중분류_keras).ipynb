```python
import tensorflow as tf
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use("ggplot")

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('data/softmax.txt', sep=' ', header=None)
df
#0, 1 특성데이터, 2, 3, 4 라벨
```

```python
x_data = df.iloc[:,0:2].values
y_data = df.iloc[:,2:].values
```

```python
#최적의 초기값 알아서 생성
dense = Dense(units = 3, input_dim = 2, activation = 'softmax') #w : 2 X 3, b : 3 xabier 객체
model = Sequential([dense])
model.compile(loss='categorical_crossentropy', optimizer=Adam(0.1), metrics=['acc']) #로그함수 적용/ 정확도 출력
h = model.fit(x_data, y_data, epochs = 1000)
```

```python
w, b = dense.get_weights()
```

```python
def resultFn(dt):
    n = int( model.predict( dt ).argmax(axis = 1) )
    grade = ['A', 'B', 'C']
    return grade[n]
```
