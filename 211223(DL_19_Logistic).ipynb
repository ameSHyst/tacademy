```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")

from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.keras.optimizers import Adam
```

```python
#공부한 시간, 출석 일수
x_data = np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]] )
y_data = np.array( [[0],[0],[0],[1],[1],[1]] )
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
# xavier, he (최적의 초기화 알고리즘)
initX = tf.initializers.GlorotUniform() #xavier
w = tf.Variable( initX( shape = [2,1] ) )
b = tf.Variable( initX( shape = [1] ) )
```

```python
def compute_loss():
    z = tf.matmul(x, w) + b
#     hx = tf.nn.sigmoid(z) #0 ~ 1
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits(labels = y, logits = z)
    cost = tf.reduce_mean( cost_i )
    #cost가 미분 불가하기 때문에 log함수 적용
    #  hx = tf.matmul( x, w) + b
    #  cost = tf.reduce_mean( (hx-y)**2) -> 연속데이터 경우
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize(compute_loss, var_list = [w, b] )
    print(i, 'cost : ', compute_loss().numpy() )
```

```python
#예측함수 생성
def hxfn(x_data):
    xd = tf.constant(x_data, tf.float32)
    z = tf.matmul(xd, w) + b
    hx = tf.nn.sigmoid(z)
    return (hx.numpy() > 0.5) + 0 #합격(1)/ 불합격(0) 출력
#     return z.numpy(), hx.numpy()
```
