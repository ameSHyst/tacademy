```python
import tensorflow as tf
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use("ggplot")

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('ddata/pima-indians-diabetes.data.csv')
df
```

```python
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1]].values
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_data)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y_data, test_size=0.3, 
                                                    stratify = y_data) #true, false 케이스 균일하게 선정
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.2, 
                                                    stratify = y_train) #true, false 케이스 균일하게 선정
```

```python
dense = Dense(units = 1, input_dim = 8, activation = 'sigmoid') #w : 8 X 1, b : 1 xabier 객체
model = Sequential([dense])
model.compile(loss='binary_crossentropy', optimizer=Adam(0.1), metrics=['acc']) #로그함수 적용/ 정확도 출력
h = model.fit(x_train, y_train, epochs=1000, validation_data = (x_valid, y_valid)) #검증 데이터셋 cost/acc도 출력
```

```python
h.history.keys()
```

```python
plt.plot(h.history['loss'])
plt.plot(h.history['val_loss'])
plt.show()
```

```python
plt.plot(h.history['acc'])
plt.plot(h.history['val_acc'])
plt.show()
```

```python
model.evaluate(x_train, y_train)
#테스트 손실값, 테스트 정확도
```

```python
model.evaluate(x_test, y_test)
```
