# 알렉사 사이트 

https://www.alexa.com/topsites

1) request 모듈 이용 페이지 요청-> html 페이지를 텍스트로 저장
2) BeautifulSoup 이용 html 페이지 -> soup 개체
3) 필요한 요소만 리스트로 저장
4) 리스트 -> 데이터프레임
5) 데이터프레임 -> csv

````python
# request 모듈 이용 페이지 요청-> html 페이지를 텍스트로 저장
url = 'https://www.alexa.com/topsites'
res = requests.get(url)
# 200 이면 성공
print(res)
# html 페이지를 텍스트로 저장
html_str = res.text
print(type(html_str))
print(html_str)
````

````python
#BeautifulSoup 이용 html 페이지 -> soup 개체
#BeautifulSoup(html페이지소스, 파서(html,parser/lxml/xml))
soup = BeautifulSoup(html_str, 'html.parser')
print(type(soup))
print(soup)
````

# soup 필터링 메서드

find(태그/클래스/아이디) -> 첫번째 요소
find_all(class_='클래스명') -> 전체 요소 클래스
find_all('태그명', {'class':'클래스명'}) -> 전체 태그 요소

select(css선택자) -> 리스트
select_one(css선택자) -> 첫번째 요소

````python
# h1 태그 안의 소스 핸들링 - find 이용
print(soup.find('h1')) #h1 태그를 찾아줘
print('='*50)
print(soup.find(class_ = 'page-title-text')) # class값이 page-title-text인 소스를 찾아줘
print('='*50)
print(soup.find(class_ = 'container'))
````

````python
# h1 태그 안의 소스 핸들링 - select_one 이용
# select_one('css선택자')
# css선택자는 클래스인 경우 ,클래스명 / 아이디는 #아이디명
print(soup.select_one('h1')) #h1 태그를 찾아줘
print('='*50)
print(soup.select_one('.page-title-text')) # class값이 page-title-text인 소스를 찾아줘
print('='*50)
print(soup.select_one('.container'))
````

````python
# h1 태그 안의 소스 핸들링 - select 이용 -> 리스트
print(soup.select('h1')) #h1 태그를 찾아줘
print('='*50)
print(soup.select('.page-title-text')) # class값이 page-title-text인 소스를 찾아줘
print('='*50)
print(soup.select('.container')) # class값이 container인 소스 전체를 찾아서 리스트에 저장해줘

# print(soup.find_all(class_='container')) 
container_list = soup.select('.container')
print('container_list 전체갯수는?', len(container_list))
print(type(container_list))
# 텍스트만 출력해줘 
# 요소.text
for item in container_list:
    print(item.text)
````

````python
# 필요한 요소만 리스트로 저장
# 필터링 요소 찾기 - 태그 위주
# 클래스 값이 'tr site-listing'인 태그를 모두 찾아서 리스트에 저장해줘
div_list = soup.find_all(class_ = 'tr site-listing')
print('='*50)
print(len(div_list)) # 전체 길이
print('='*50)
print(div_list) #전체
print('='*50)
print(div_list[0]) #첫번째 소스
print('='*50)
print(div_list[-1]) #마지막 소스
````

````python
#첫번째 요소에 필요한 텍스트만 추출

#첫 번째 div 태그 안에 텍스트를 찾아서 저장해줘
rank = div_list[0].find('div').text
#첫 번째 a 태그 안에 텍스트를 찾아서 저장해줘
site_name = div_list[0].find('a').text
#첫 번째 a 태그 안에 href 속성값을 찾아서 저장해줘
#[속성], get(속성)
#site_url = div_list[0].find('a')['href']
site_url = div_list[0].find('a').get('href')
#두 번째 'td right' 클래스의 텍스트를 찾아서 저장해줘
page_view = div_list[0].find_all(class_ = 'td right')[1].text

print(rank, site_name, site_url, page_view)
````

````python
#모든 요소를 찾아서 2차원 리스트에 저장
alexa_list = []
for row in div_list:
    rank = row.find('div').text
    site_name = row.find('a').text
    site_url = row.find('a').get('href')
    page_view = row.find_all(class_ = 'td right')[1].text
    #1차원 리스트로 생성한 후 다시 리스트에 삽입
    alexa_list.append([rank, site_name, site_url, page_view])
````

````python
# 리스트 -> 데이터프레임
# 데이터프레임 -> csv 저장
df_alexa = pd.DataFrame(alexa_list, columns= ['rank', 'site_name', 'site_url', 'page_view'])
df_alexa.sample(5)
````

# 코드 개선 -> 함수로

````python
base_page_url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page='
````

````python

def save_df(base_page):
    # 각 페이지의 공통 주소 부분 
    base_page = base_page_url
    kid_list = []
    base_url = 'https://movie.naver.com'
    page=0

    # 무한 루프 
    while(True):
        # 실제 페이지 주소 생성 
        url = base_page+str(page+1)
        res = requests.get(url)
        # 200 이면 성공 
        print(res, url) 
        # html 페이지를 텍스트로 저장 
        html_str = res.text

        # soup 객체화    
        soup = BeautifulSoup(html_str, 'html.parser')

        # 제목, 평점, 리뷰, 링크, 날짜와 작성자 텍스트 추출 
        td_list = soup.select('.title')
        a_list = soup.select('.author')

        # 모든 요소를 찾아서 2차원 리스트에 저장
        for j in range(len(a_list)):
            title = td_list[j].select_one('a').text
            score = td_list[j].select_one('em').text
            review = td_list[j].select_one('br').next_sibling.strip()
            link = base_url + td_list[j].select_one('a')['href']
            writer = a_list[j].text
            date = a_list[j].next_sibling.next_sibling

            # 1차원 리스트로 생성한 후 다시 리스트에 삽입      
            kid_list.append([title, score, review, link, writer, date])  


        #다음 버튼 찾기
        next_btn = soup.find_all(class_='pg_next')
        #다음 버튼이 없다면 종료
        if len(next_btn) == 0:
            break


        #페이지 증감
        page += 1

    print('테스트 종료')
    df = pd.DataFrame(kid_list, columns=['title', 'score', 'review', 'link', 'writer', 'date'])
    
    return df
````

````python
import os
import sys
import urllib.request
````

````python
def book_api(book_name, max_num):
    
    encText = urllib.parse.quote(book_name)
    sep = round(int(max_num)/50)
    temp = []
    
    for a in range(0, sep):
        a = a * 50 + 1
        url = "https://openapi.naver.com/v1/search/book.json?start={}&display=50&query=".format(a) + encText # json 결과
    
    
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id",client_id)
        request.add_header("X-Naver-Client-Secret",client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
    
        # 페이지가 정상적으로 접근되어진다면 저장 그렇지 않으면 에러메세지와 코드 출력 
        if(rescode==200):
            response_body = response.read()
#     print(response_body.decode('utf-8'))
        else:
            print("Error Code:" + rescode)
        
        # json.loads(response_body) => 딕셔너리 => 리스트 
        book_list = json.loads(response_body)['items']
        books = pd.DataFrame(book_list)
        temp.append(books)        
        
    df = pd.concat(temp, axis=0).reset_index(drop=True)
    
    return df
````
