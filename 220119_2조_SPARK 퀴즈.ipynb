## . births.txt 파일을 이용하여 구하시요

```python
b= sc.textFile('../data/births.txt')
b.collect()
```

1) 남아수의 합과 평균을 구하시요

```python
b.map(lambda v : v.split(',') ).map(lambda v: int(v[1])).sum(),b.map(lambda v : v.split(',') ).map(lambda v: int(v[1])).mean()
```

2) 여아수의 합과 평균을 구하시요

```python
b.map(lambda v : v.split(',') ).map(lambda v: int(v[2])).sum(),b.map(lambda v : v.split(',') ).map(lambda v: int(v[2])).mean()
```

3) 남아수가 높은 탑 5를 구하시요

```python
b.map(lambda v : v.split(',') ).map(lambda v: int(v[1])).sortBy(lambda v:v ,ascending=False).collect()[0:5]
```

```python
b.map(lambda v : v.split(',') ).map(lambda v: int(v[1])).sortBy(lambda v:v ,ascending=False).take(5)
```

4) 전체(남아여아의총수)중 남아수의 비율을 구하시요.

```python
b.map(lambda v : v.split(',') ).sortBy(lambda x:int(x[1]), ascending=False).take(5)
```

```python
# 남아 수 
m = b.map(lambda v : v.split(',') ).map(lambda v: int(v[1])).sum()
# 여아수
f = b.map(lambda v : v.split(',') ).map(lambda v: int(v[2])).sum()



#전체수 
tot = m + f
# 남아  전체 
print(  (m/tot)*100,"%" ,sep="")
```
