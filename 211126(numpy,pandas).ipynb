-컬럼 값을 특정 조건에 따라 교체하기

ex)
0 -> mon, 1-> tue, ...
남자 -> 1, 여자 -> 2 ...
강서구 방화동 -> 방화동

for + if 등의 제어문
map({old1:new1, old2:new2 ... }) 
replace({old1:new1, old2:new2 ... })
replace([old1, old2 ...], [new1,new2 ...] })
apply()

```python
#df2['d6'] 컬럼 복제
df2['d7'] = df2['d6']
df2
```

```python
#d7 컬럼에서 mon, tue, wed, ...
#replace({old1:new1}) 
#원본 컬럼에 적용
df2['d7'] = df2['d7'].replace({0:'mon', 1:'tue', 2:'wed', 3:'thu', 4:'fri', 5:'sat', 6:'sun'})
df2
```

```python
#apply() 함수를 이용해서 값 교체
#apply() 에 적용되는 함수 정의
def change_day(x) :
    week_day = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']
    if 0 <= x <= 6:
        return week_day[x]
    else:
        return '잘못된 입력'
```

```python
change_day(0), change_day(5)
```

```python
df2['d6'].apply(change_day)
```

```python
#d6 컬럼의 apply() 적용해서 새로운 컬럼 생성
df2['d8'] = df2['d6'].apply(change_day)
df2
```

-NaN 값이 들어있는 행 삭제하기

데이타프레임이름.dropna(how="any", inplace=False/True) : 결측값이 하나라도 있다면 삭제
데이타프레임이름.dropna(how="all", inplace=False/True) : 결측값이 모두 있다면 삭제

```python
# 결측값이 있는 데이타프레임 생성 
df = pd.DataFrame(np.arange(1,21).reshape(5,4))
df.index = ['row'+str(i) for i in range(1,6)]
df.columns = ['col'+str(i) for i in range(1,5)]
df.loc['row1','col2'] = np.nan
df.loc['row4','col1'] = np.nan
df.loc['row5','col1'] = np.nan
df.loc['row5','col2'] = np.nan
df.loc['row5','col3'] = np.nan
df.loc['row5','col4'] = np.nan
df
```

```python
#하나라도 결측치가 있다면 삭제
df.dropna(how='any')
```

```python
#결측치가 모두 있다면 삭제
df.dropna(how='all', inplace=True)
```

```python
df
```

```python
# 결측값이 있는 데이타프레임 생성 
df = pd.DataFrame(np.arange(1,21).reshape(5,4))
df.index = ['row'+str(i) for i in range(1,6)]
df.columns = ['col'+str(i) for i in range(1,5)]
df.loc['row1','col2'] = np.nan
df.loc['row4','col1'] = np.nan
df.loc['row5','col1'] = np.nan
df.loc['row5','col2'] = np.nan
df.loc['row5','col3'] = np.nan
df.loc['row5','col4'] = np.nan
df
```
