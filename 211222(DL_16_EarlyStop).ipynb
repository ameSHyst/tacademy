```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("ggplot")

from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

### early stop
    과적합 방지
    학습 횟수 최대한/ 학습이 완료되면 학습 종료
    
    patience : 기다리는 횟수
    min_delta : cost 값의 차이 (0.01, 0.001)

```python
df = pd.read_csv('data/test.csv',header=None)
df.columns = ['q1','q2','mid','final']
df
```

```python
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1]].values
```

```python
x_data
```

```python
y_data
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(tf.random.uniform( [3, 1] )) 
b = tf.Variable(tf.random.uniform( [1] ))
```

```python
#백문이불여일'타'
#코딩은 학습이 아닌 연습
```

```python
def compute_loss():
    hx = tf.matmul(x, w) + b
    cost = tf.reduce_mean( (hx-y)**2 )
    return cost
```

```python
patience = 20
min_delta = 0.01 #이전 cost와 현재 cost를 비교해서 20번 동안 min_delta보다 작은 상태가 유지되면 학습 종료하고 빠져나감
hist_cost = [] #10, 9, 8, ....
pcnt = 0
optimizer = Adam(0.1)

for i in range(10000):
    optimizer.minimize(compute_loss, var_list = [w, b])
    c = compute_loss().numpy()
    hist_cost.append(c)
    print(i, 'cost : ', c)
    if i > 0:
        if hist_cost[i-1] - hist_cost[i] > min_delta:
            pcnt = 0
        else:
            pcnt += 1

        if pcnt >= patience:
            print('early stop')
            break
```

#### 케라스 버전

```python
dense = Dense(units = 1, input_dim = 3) #xavier, he 알고리즘 추후 소개
model = Sequential( [dense] )
model.compile(loss = 'mse', optimizer = Adam(0.1))
earlyStop = EarlyStopping(monitor = 'loss', min_delta = 0.01, patience = 100)
h = model.fit(x_data, y_data, epochs = 10000, callbacks = [earlyStop]) #callback : 특정 기능 수행
```
