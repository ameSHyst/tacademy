```python
from numpy import array
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, SimpleRNN, GRU, TimeDistributed
```

#### Many to One

```python
x = array([[1,2,3], [2,3,4], [3,4,5], [4,5,6]]) #[1, 2, 3]으로 '4' 예측, [2, 3, 4]로 '5' 예측, ...
y = array([4,5,6,7])
print('x shape : ', x.shape) # (4,3)
print('y shape : ', y.shape) # (4,)
print(x)
```

```python
x = x.reshape(4, 3, 1) # (batch size, sequence length, feature size)
x
#       □
# □ □ □
# □ □ □
# □ □ □
# □ □ □
```

```python
model = Sequential()                    
                 #이 값에 따라 rnn 셀의 크기가 결정됨
                 #    ↓                    sequence length, feature size
model.add( SimpleRNN( 8, activation='relu', input_shape=(3, 1) ) )
# model.add( Dense( units=16 ) )
model.add( Dense( units=1 ) )
model.compile( loss='mse', optimizer='adam' )
model.fit(x, y, epochs=1000)
```

```python
model.layers
```

```python
rnnW = model.layers[0].get_weights()
rnnW
```

#### Many to Many

```python
import numpy as np
```

```python
X = list()
Y = list()
X = [x for x in range(5, 301, 5)]
Y = [y for y in range(20, 316, 5)]

X = np.array(X).reshape(20, 3, 1)
Y = np.array(Y).reshape(20, 3, 1)
```

```python
X
```

```python
Y
```

```python
model = Sequential()
model.add(LSTM(64, activation='relu', input_shape=(3, 1), return_sequences=True) )
model.add(TimeDistributed( Dense(1) ) )
#각 dense를 timestep별로 matmul하겠다
model.compile(loss='mse', optimizer='adam')
model.fit(X, Y, epochs=1000)
```

```python
arr = np.array( [300, 305, 310] )
model.predict( arr.reshape(1, 3, 1) )
```
