```python
import tensorflow as tf
import numpy as np
import pandas as pd

from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv('data/test.csv', header=None)
df.columns=['q1', 'q2', 'mid', 'final']
df
```

```python
x_data = df.iloc[:, :-1].values #특성데이터 q1+q2+mid
y_data = df.iloc[:, [-1]].values #라벨 final
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable(tf.random.uniform( [3, 1] )) #x와 w를 matmul / [특성데이터 갯수, 라벨 갯수]
b = tf.Variable(tf.random.uniform( [1] )) #라벨 갯수
```

```python
def compute_loss():
    hx = tf.matmul(x, w) + b
    cost = tf.reduce_mean( (hx-y)**2 )
    return cost
```

```python
optimizer = Adam(0.1)
for i in range(1000):
    optimizer.minimize( compute_loss, var_list = [w, b] )
    print(i, 'cost : ', compute_loss().numpy() )
```

```python
def hxFn(x_data):
    xd = tf.constant(x_data, tf.float32)
    hx = tf.matmul(xd, w) + b
    return hx.numpy()
```

```python
#q1 = 80, q2 = 90, mid = 90인 경우의 final 예측
hxFn( [[80,90,90]] )
```

```python
hxFn(x_data)
```
