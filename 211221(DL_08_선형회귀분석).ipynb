```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
df = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
df
#접기 : o
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
plt.scatter(x_data, y_data)
plt.show()
```

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
w = tf.Variable( tf.random.uniform([1]) ) #임의의 값 부여
b = tf.Variable( tf.random.uniform([1]) )
```

```python
def compute_cost():
    hx = w*x + b
    cost = tf.reduce_mean((hx-y)**2)
    return cost
```

```python
optimizer = Adam(learning_rate = 0.01)
hist_cost = []
for i in range(2000):
    optimizer.minimize(compute_cost, var_list=[w, b]) 
    #compute_cost 자리 -> 함수의 주소 요구하는 인자이기 때문에 그냥 cost 대신 함수 입력
    c = compute_cost().numpy()
    hist_cost.append(c)
    print(i, 'cost = ', c)
```

```python
def hxFn(x_data):
    # xd = tf.constant(x_data)
    # hx = w*xd + b
    hx = w*x_data + b
    return hx.numpy()
```

```python
#자동차 속도가 10인 경우 제동거리
hxFn(10)
```

```python
#자동차 속도가 10, 14인 경우 제동거리
hxFn( [10, 14] )
```

```python
plt.plot(hist_cost)
plt.show()
```
