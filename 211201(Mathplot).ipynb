## plot 객체 생성하기 - 파이플롯 스타일

- figVar = plt.figure(figsize=(가로,세로)) : 가로*세로 크기에 해당하는 fig 생성 
- figVar, plt.plot(xData,yData,label=labelName) : figure에 플랏 생성하기 
- plt.xlabel(X축)
- plt.ylabel(Y축)
- plt.title(제목)
- plt.legend() : 범례 표시 
- plt.show() : plot 다시 그리기 

````python
#figVar = plt.figure(figsize=(가로,세로)) : 가로*세로 크기에 해당하는 fig 생성
#                            [가로, 세로]
fig1 = plt.figure(figsize=(5,5))
print(type(fig1))

#데이터 생성
x = np.arange(-5,5)
y = 2*x

#그래프화
#plt.plot(xData,yData,label=labelName)
#label은 그래프 범례
plt.plot(x,y,label='직선 2*x')

#그리드 표시
plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')

#범례 표시
plt.legend()

#그래프 제목 설정
plt.title('직선 그래프')

#그래프 다시 그리기(최종 정리/ 다른 셀에서 따로 실행은 불가/ 항상 마지막에 위치해야)
plt.show()
````

# 제목, 범례, 그래프 속성 

- https://matplotlib.org/stable/tutorials/text/text_intro.html

### fig 객체 속성 
```
 plt.figure((옵션속성=값)
       - fig변수 = plt.figure(옵션속성=값)
       - figsize는 ~10. 그래프전체틀 크기 
       - facecolor= 영문색상명이나 16진수색상값 # 배경색상
       - frameon=True/False # 테두리프레임 Show/Hide
       - edgecolor = 영문색상명이나 16진수색상값 #테두리색상
```

### 그래프제목 속성

```
plt.title('그래프제목', 
          fontdict={'폰트속성':값, '폰트속성':값 ...}, 
          loc=위치예약어, # left/right/center
           pad=숫자)

예) fontdict
fontdict={'fontsize':30, 
            'color':'white', 
            'fontweight':'bold',
             'fontfamily':'Georgia'}         
```

### 범례(legend) 속성 
```
plt.legend(loc=숫자나위치예약어, 
            labels=[범례1, 범례2, 범례3], 
            labelspacing=숫자,  #숫자가 높을수록 범례간격이 넓어진다. 
             facecolor=배경색상값,
             borderaxespad = 숫자, #숫자가 높을수록 바깥여백이 커진다.
             edgecolor=테두리색상
             )
```

```
loc 예약어와 코드값

        'best'            0
        'upper right'     1
        'upper left'      2
        'lower left'      3
        'lower right'     4
        'right'           5
        'center left'     6
        'center right'    7
        'lower center'    8
        'upper center'    9
        'center'          10
```

````python
fig4 = plt.figure(figsize=(20,7), facecolor='yellowgreen', #그래프 배경 색상
                  frameon=True, #figure 테두리 표시 
                  edgecolor='black') #테두리 색상

x = np.linspace(-5,5,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label='sin', 
         color='r', linewidth=5, 
         marker='o', markersize=10,  markerfacecolor='b')

plt.plot(x,y2, 'gv--', label='cos')

#제목 설정
#fontdict={옵션:값..}
#loc는 위치 값 -> left, center, right
#pad는 여백
plt.title('sin, cos 그래프', fontdict={'fontsize':30, 'color': 'darkgreen', 'fontweight': 'bold'},
         loc='right', pad=50)

plt.grid()
plt.xlabel('x axis')
plt.ylabel('y axis')

#범례 설정
#edgecolor = 테두리 색상
#facecolor = 배경 색상
#ncol 숫자값이 컬럼 갯수
#borderaxespad = 범례 밖의 여백
plt.legend(loc='lower right', facecolor='yellow', borderaxespad=3, edgecolor='crimson', ncol=2)

#이미지 파일로 저장하기
#plt.savefig(fileURL)
plt.savefig('output/fig4.png')

plt.show()
````

# add_subplot 방식의 그래프 

```
fig_a = plt.figure(frameon=True, figsize=[8, 10])

ax = fig_a.add_subplot()

ax.plot( x,y, label='y=2*x')
```

````python
fig = plt.figure(figsize=[5, 5], frameon=True, dpi=100)

x = np.arange(-5,5)
y = 2*x

ax = fig.add_subplot()

ax.plot(x,y,label='y=2*x')
ax.legend(  loc=1 , 
              facecolor='GhostWhite',
              borderaxespad = 2, 
              edgecolor='darkblue',
              fontsize = 15
           )

ax.set_title("직선 그래프", 
          fontdict={'fontsize':30, 'color':'darkgreen', 'fontweight':'bold'}, 
          loc='center', pad=50)

ax.set_xlabel('x axis')
ax.set_ylabel('y axis')

ax.set_xticks([i for i in range(-5,5)])
ax.set_yticks([i for i in range(-10,10)])

ax.set_xlim([-5,5])
ax.set_ylim([-9,9])

ax.grid()
plt.show()
````

## barh plot

- figure 변수 생성 
- subPlots() 가변 
``` 
figure, subplotVar = plt.subplots() 
```
- x, y 데이타 생성
- subplotVar.barh(dataList, Value, color=color Name, widtgh=float)
- y축의 눈금을 데이타로 지정하기 
```
 subplotVar.set_yticks(xData)
 subplotVar.set_yticklabels(xDataList)
```

````python
#x 데이터
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim', 'Kyle')
#총 데이터
y_pos = np.arange(len(people))

performance = np.random.rand(len(people)) 

# 그래프 전체 변수 설정 
fig16 = plt.figure(figsize=(4,4))

# 변수 ax 그래프 변수 
fig16, ax = plt.subplots()

# 변수.barh(x, y, 옵션들)
ax.barh(y_pos, performance, color='green')

# y축의 라벨 설정 
ax.set_yticks(y_pos)
ax.set_yticklabels(people)

# x축 이름 설정 
ax.set_xlabel('Performance')

# 제목 설정 
ax.set_title('Today\' Man ')

plt.show()
````

## Pie Plot

````python
# Label + Value 
label_list = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10] #합이 꼭 100 아니어도 나옴 (전체 결과 안의 비율)
fig_p1 = plt.figure(figsize=(5,5))
fig_p1, plt.pie(sizes, labels=label_list, autopct='%.2f%%')
                                #퍼센트 표시 없애려면 '%.2f'
                                                    #'%d%%'
plt.show()
````

# 판다스의 데이타프레임 그래프

### 데이타프레임
데이타프레임명.plot( kind=그래프형태, figsize=(가로,세로), title=제목)
그래프형태는?

    - 'line' : line plot (default)
    - 'bar' : vertical bar plot

````python
data = {'kor':[40, 55, 66, 77, 34],
        'math':[70, 55, 66, 20, 40],
        'eng':[80, 95, 100, 77, 90],
        'sci':[90, 45, 77, 90, 67]}
        
name_list = ['고길동','김희동','신윤지','이민호','박준희']    

df = pd.DataFrame(data, index=name_list)
df
````

````python
#세로막대 데이터프레임 전체 그래프
df.plot(kind='bar', title='중간고사 성적', figsize=(8,5))
#x축 설정
plt.tick_params(axis='x', labelsize=20, colors='black', width='3', rotation=45)
plt.grid()
````

````python
#가로막대 데이터프레임 전체 그래프
df.plot(kind='barh', title='중간고사 성적', figsize=(8,5))
plt.tick_params(axis='y', labelsize=15, colors='blue', width='3', rotation=45)
````

# 학생별로 막대그래프 출력

````python
df.index
````

````python
#2행 3열 레이아웃
plt.figure(figsize=(10,12))

plt.subplot(231)
df.iloc[0].plot(kind='bar', title=df.index[0], color=['r','b','g','c']);

plt.subplot(232)
df.iloc[1].plot(kind='bar', title=df.index[1], color=['r','b','g','c']);

plt.subplot(233)
df.iloc[2].plot(kind='bar', title=df.index[2], color=['r','b','g','c']);

plt.subplot(234)
df.iloc[3].plot(kind='bar', title=df.index[3], color=['r','b','g','c']);

plt.subplot(235)
df.iloc[4].plot(kind='bar', title=df.index[4], color=['r','b','g','c']);

#그래프 전체 제목 설정(supertitle)
plt.suptitle('학생별 중간고사 성적')
plt.axis('auto')
plt.show()
````

````python
#for문을 이용해서 그래프 출력
plt.figure(figsize=(10,12))

for i in range(4):    
    #221, 222, 223, 224
    plt.subplot(221+i)
    df.iloc[i].plot(kind='bar', title=df.index[i], color=['r','b','g','c']);

plt.suptitle('학생별 중간고사 성적')
plt.show()
````
