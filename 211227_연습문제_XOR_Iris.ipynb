#### Iris - [5.1, 3.5, 1.4, 0.2] 종분류 예측/ 정확도
#### 딥러닝(텐서, 케라스)

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
irisDF = pd.DataFrame( iris['data'])
irisDF.columns = iris['feature_names']
irisDF['species'] = iris['target']
irisDF
```

```python
pd.get_dummies( irisDF['species'])
```

```python
x_data = irisDF.iloc[:, :-1]
y_data = pd.get_dummies( irisDF['species'])
```

#### Tensor ver.

```python
x = tf.constant(x_data, tf.float32)
y = tf.constant(y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform() #32의 배수로?

w1 = tf.Variable( initX( shape = [4, 64] ) )
b1 = tf.Variable( initX( shape = [64] ) )

w2 = tf.Variable( initX( shape = [64, 3] ) )
b2 = tf.Variable( initX( shape = [3] ) )
```

```python
def compute_loss():
    z1 = tf.matmul(x, w1) + b1
    hx1 = tf.nn.softmax(z1)
    z = tf.matmul(hx1, w2) + b2 #최종 z값
    
    cost_i = tf.nn.softmax_cross_entropy_with_logits( logits = z, labels = y )
    cost = tf.reduce_mean(cost_i)
    return cost
```

```python
optimizer = Adam( learning_rate = 0.1)
for i in range(1000):
    optimizer.minimize (compute_loss, var_list = [w1, w2, b1, b2])
    print(i, compute_loss().numpy() )
```

```python
def hxfn( x_data ):
    xd=tf.constant(x_data, tf.float32)
    z1=tf.matmul(xd,w1)+b1
    
    hx1=tf.nn.softmax(z1)
    z=tf.matmul(hx1,w2)+b2
    
    hx = tf.nn.softmax( z )
    return (hx.numpy()>0.5)+0
```

```python
hxfn(x_data) #4x2  2x64 = 4x64  64x32  4x32  32x1 = 4x1
```

```python
hxfn( [[5.1, 3.5, 1.4, 0.2]]).argmax(axis=1)
```

#### Accuracy

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score
accuracy = accuracy_score(y_data, hxfn(x_data))
accuracy
```

#### Keras ver.

```python
modelNN = Sequential()

modelNN.add( Dense( units=64, input_dim=4 ,activation="softmax") ) #4x64, b=64 
modelNN.add( Dense( units=3, input_dim=64, activation="softmax") ) #64x1, b=3
modelNN.compile( loss="categorical_crossentropy" ,optimizer=Adam(0.1), metrics=['acc'] )
h = modelNN.fit( x_data, y_data, epochs=1000)
```

```python
(modelNN.predict(x_data) > 0.5) + 0
```

```python
modelNN.predict( [[5.1, 3.5, 1.4, 0.2]] ).argmax( axis=1)
```
